#### Preprocessing & Wrappers

```python
def make_pong_env(env_id="ALE/Pong-v5",
                  clip_rewards=True,
                  frame_stack=4,
                  screen_size=84,
                  noop_max=30,
                  render_mode="rgb_array"):
  """
  Creates and wraps the Atari Pong environment.

  Args:
      env_id (str): The Gymnasium ID for the Pong environment.
      clip_rewards (bool): Whether to clip rewards to {-1, 0, 1}.
      frame_stack (int): Number of frames to stack.
      screen_size (int): The size to resize a single dimension of the grayscale image to (e.g., 84 for 84x84).
      noop_max (int): Maximum number of no-op actions to perform at the start of an episode.
  """
  env = gym.make(env_id, obs_type='rgb', render_mode='rgb_array', frameskip=1)
  env = AtariPreprocessing(env, noop_max = noop_max, screen_size = screen_size, grayscale_obs=True, scale_obs=False, terminal_on_life_loss=False)

  if clip_rewards:
    env = ClipReward(env, min_reward=-1, max_reward=1)

  if frame_stack > 1:
    env = FrameStackObservation(env, frame_stack)

  return env
```

### Q-Network

```python
class DQN(nn.Module):
  """
    Deep Q-Network for Atari games
    Input: Stacked grayscale frames (4, 84, 84)
    Output: Q-values for each action
    """
  def __init__(self, n_actions=6):
    super(DQN, self).__init__()

    # setup the convolutions of neural nets
    self.conv1 = nn.Conv2d(4, 32, kernel_size=8, stride=4)
    self.conv2 = nn.Conv2d(32, 64, kernel_size=4, stride=2)
    self.conv3 = nn.Conv2d(64, 64, kernel_size=3, stride=1)

    conv_output_size = self.get_conv_out_size((4, 84, 84))

    #  fully connected layers for decision making
    self.fc1 = nn.Linear(conv_output_size, 512)
    self.fc2 = nn.Linear(512, n_actions)

  def get_conv_out_size(self, shape):
    """Calculate output size after convolutional layers"""
    dummy_input = torch.zeros(1, *shape)
    dummy_output = self._forward_conv(dummy_input)
    return int(np.prod(dummy_output.size()))

  def _forward_conv(self, x):
    """Forward pass thru convolutional layers only"""
    x = F.relu(self.conv1(x))
    x = F.relu(self.conv2(x))
    x = F.relu(self.conv3(x))
    return x

  def forward(self, x):
    """Full forward pass"""
    # normalize the pixel to [0,1]
    x = x.float() / 255.0

    # pass thru all layers
    x = F.relu(self.conv1(x))
    x = F.relu(self.conv2(x))
    x = F.relu(self.conv3(x))

    # flatten for FC layers
    x = x.view(x.size(0), -1)

    x = F.relu(self.fc1(x))
    q_values = self.fc2(x)

    return q_values
```

```python
def initialize_dqn(n_actions=6, device="cuda" if torch.cuda.is_available() else "cpu"):
  q_network = DQN(n_actions = n_actions)

  print(device)
  # move to device
  q_network = q_network.to(device)

  # Initialize weights manually (alternative to .apply())
  with torch.no_grad():
    # Initialize convolutional layers
    nn.init.xavier_uniform_(q_network.conv1.weight)
    nn.init.constant_(q_network.conv1.bias, 0)
    nn.init.xavier_uniform_(q_network.conv2.weight)
    nn.init.constant_(q_network.conv2.bias, 0)
    nn.init.xavier_uniform_(q_network.conv3.weight)
    nn.init.constant_(q_network.conv3.bias, 0)

    # Initialize fully connected layers
    nn.init.xavier_uniform_(q_network.fc1.weight)
    nn.init.constant_(q_network.fc1.bias, 0)
    nn.init.xavier_uniform_(q_network.fc2.weight)
    nn.init.constant_(q_network.fc2.bias, 0)

  return q_network, device
```

### DQN with visuals

```python
class DQNWithVisualization:
  def __init__(self, actions):
    self.actions = actions
    self.q_net, self.device = initialize_dqn(6)
    self.target_net, self._ = initialize_dqn(6)
    self.replay_mem = deque(maxlen=1000)
    self.epsilon = 0.9
    self.optimizer = torch.optim.Adam(self.q_net.parameters(), lr=0.0001)
    self.loss_fn = nn.MSELoss()
    self.gamma = 0.99

    # Visualization tracking
    self.episode_rewards = []
    self.episode_lengths = []
    self.losses = []

  def random_action(self):
    return random.choice(self.actions)

  def get_action(self, obs, use_epsilon=True):
    """Get action from Q-network with optional epsilon-greedy exploration"""
    obs_tensor = torch.FloatTensor(obs).unsqueeze(0).to(self.device)
    q_values = self.q_net.forward(obs_tensor)
    best_action = torch.argmax(q_values, dim=1).item()

    if use_epsilon and random.random() < self.epsilon:
      return self.random_action()
    else:
      return best_action

  def visualize_episode(self, episode_num=0, max_steps=1000, save_frames=False):
    """
    Visualize agent playing one episode
    """
    env = make_pong_env("ALE/Pong-v5", True, 4, 84, 30, render_mode="rgb_array")
    obs, info = env.reset()

    frames = []
    total_reward = 0
    step = 0

    # Create figure for real-time display
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    display(fig) # Display the figure initially

    for step in range(max_steps):
      # Get action from trained network (no exploration)
      action = self.get_action(obs, use_epsilon=False)

      # Take step
      next_obs, reward, terminated, truncated, info = env.step(action)
      total_reward += reward

      # Render and capture frame
      rgb_frame = env.render()
      if save_frames:
        frames.append(rgb_frame)

      # Update display every 10 steps for performance
      if step % 10 == 0:
        # Display game frame
        ax1.clear()
        ax1.imshow(rgb_frame)
        ax1.set_title(f'Episode {episode_num} - Step {step}\\nReward: {total_reward:.1f}')
        ax1.axis('off')

        # Display processed observation (first frame of stack)
        ax2.clear()
        ax2.imshow(obs[0], cmap='gray')
        ax2.set_title('Processed Frame (Grayscale)')
        ax2.axis('off')

        plt.tight_layout()
        # Update the displayed figure
        display(fig)


      obs = next_obs

      if terminated or truncated:
        break

    env.close()
    plt.close(fig) # Close the figure after the episode

    print(f"Episode finished! Steps: {step+1}, Total Reward: {total_reward}")

    if save_frames:
      return frames

    return total_reward, step+1

  def create_training_plots(self):
    """Create plots showing training progress"""
    if len(self.episode_rewards) > 0:
      fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

      # Episode rewards
      ax1.plot(self.episode_rewards)
      ax1.set_title('Episode Rewards Over Time')
      ax1.set_xlabel('Episode')
      ax1.set_ylabel('Total Reward')
      ax1.grid(True)

      # Episode lengths
      ax2.plot(self.episode_lengths)
      ax2.set_title('Episode Lengths Over Time')
      ax2.set_xlabel('Episode')
      ax2.set_ylabel('Steps')
      ax2.grid(True)

      # Loss over time
      if len(self.losses) > 0:
        ax3.plot(self.losses[-1000:])  # Show last 1000 losses
        ax3.set_title('Training Loss (Last 1000 updates)')
        ax3.set_xlabel('Update Step')
        ax3.set_ylabel('MSE Loss')
        ax3.grid(True)

      # Epsilon decay
      episodes = list(range(len(self.episode_rewards)))
      epsilons = [max(0.1, 0.9 * (0.995 ** (ep // 10))) for ep in episodes]
      ax4.plot(episodes, epsilons)
      ax4.set_title('Epsilon Decay')
      ax4.set_xlabel('Episode')
      ax4.set_ylabel('Epsilon')
      ax4.grid(True)

      plt.tight_layout()
      plt.show()

  def execute_dqn_with_visualization(self, max_episodes=25, visualize_every=5):
    """ Execute DQN training with periodic visualization """
    env = make_pong_env("ALE/Pong-v5", True, 4, 84, 30)
    TARGET_UPDATE_FREQUENCY = 5

    self.target_net.load_state_dict(self.q_net.state_dict())

    for episode in range(max_episodes):
      obs, info = env.reset()
      episode_reward = 0
      episode_length = 0

      for t in range(1000):
        # get next action
        action = self.get_action(obs, use_epsilon=True)

        # take action
        next_obs, reward, terminated, truncated, info = env.step(action)
        done = terminated or truncated
        episode_reward += reward
        episode_length += 1

        # store experience
        if len(self.replay_mem) == self.replay_mem.maxlen:
          self.replay_mem.popleft()
        self.replay_mem.append((obs, action, reward, next_obs, done))

        # training
        if len(self.replay_mem) > 30:
          loss = self.train_step()
          self.losses.append(loss)

        obs = next_obs

        # update target network
        if t%TARGET_UPDATE_FREQUENCY == 0:
          self.target_net.load_state_dict(self.q_net.state_dict())

        if done:
          break

      # store episode statistics
      self.episode_rewards.append(episode_reward)
      self.episode_lengths.append(episode_length)

      # decay epsilon
      if episode%10 == 0:
        self.epsilon = max(0.1, self.epsilon * 0.995)
        print(f"Episode {episode}: Reward={episode_reward:.1f}, "
                      f"Length={episode_length}, Epsilon={self.epsilon:.3f}")

        if episode > 0:
          self.create_training_plots()

      # visualize agent playing
      if episode % visualize_every == 0 and episode > 0:
        print(f"\n🎮 Visualizing Episode {episode}")
        self.visualize_episode(episode_num=episode, max_steps=500)


    env.close()

    # final visualization
    print("\n🎯 Final trained agent performance:")
    self.visualize_episode(episode_num="Final", max_steps=1000)


  def train_step(self):
    batch_size = 30
    replay_mem_list = list(self.replay_mem)
    minibatch = random.sample(replay_mem_list, batch_size)

    states = torch.FloatTensor([transition[0] for transition in minibatch]).to(self.device)
    actions = torch.LongTensor([transition[1] for transition in minibatch]).to(self.device)
    rewards = torch.FloatTensor([transition[2] for transition in minibatch]).to(self.device)
    next_states = torch.FloatTensor([transition[3] for transition in minibatch]).to(self.device)
    dones = torch.BoolTensor([transition[4] for transition in minibatch]).to(self.device)
    non_terminal_mask = ~dones

    # current Q-values
    curr_q_values = self.q_net(states)
    curr_q_selected = curr_q_values.gather(1, actions.unsqueeze(1)).squeeze(1)

    # target Q-values
    with torch.no_grad():
      next_q_values = self.target_net(next_states)
      max_next_q = torch.max(next_q_values, dim=1)[0]
    target_q_values = rewards + (self.gamma * max_next_q * non_terminal_mask)

    # loss calculation & backpropagation
    loss = self.loss_fn(curr_q_selected, target_q_values)

    self.optimizer.zero_grad()
    loss.backward()
    self.optimizer.step()

    return loss.item()

  def save_video(self, filename="pong_agent.mp4", fps=30, max_steps=1000):
    """ Save a video of the agent playing """
    try:
      import moviepy.editor as mp

      frames = self.visualize_episode(max_steps=max_steps, save_frames=True)

      if frames:
        # Convert frames to video
        clip = mp.ImageSequenceClip(frames, fps=fps)
        clip.write_videofile(filename, verbose=False, logger=None)
        print(f"✅ Video saved as {filename}")

        # Display video in Colab
        from IPython.display import HTML
        video = open(filename, 'rb').read()
        video_encoded = base64.b64encode(video).decode('ascii')
        video_tag = f'''
        <video width="640" height="480" controls>
        <source src="data:video/mp4;base64,{video_encoded}" type="video/mp4">
        </video>
        '''
        display(HTML(video_tag))
      else:
        print("❌ No frames captured for video")

    except ImportError:
      print("❌ moviepy not available. Install with: !pip install moviepy")
    except Exception as e:
      print(f"❌ Error creating video: {e}")

  def compare_random_vs_trained(self, max_steps=500):
    """ Compare random agent vs trained agent side by side """
    env1 = make_pong_env("ALE/Pong-v5", True, 4, 84, 30, render_mode="rgb_array")
    env2 = make_pong_env("ALE/Pong-v5", True, 4, 84, 30, render_mode="rgb_array")

    obs1, _ = env1.reset()
    obs2, _ = env2.reset()

    reward1, reward2 = 0, 0

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    display(fig) # Display the figure initially


    for step in range(max_steps):
      # Random agent
      action1 = env1.action_space.sample()
      next_obs1, r1, term1, trunc1, _ = env1.step(action1)
      reward1 += r1

      # Trained agent
      action2 = self.get_action(obs2, use_epsilon=False)
      next_obs2, r2, term2, trunc2, _ = env2.step(action2)
      reward2 += r2

      # Update display every 15 steps
      if step % 15 == 0:
        # Random agent display
        ax1.clear()
        ax1.imshow(env1.render())
        ax1.set_title(f'Random Agent\\nReward: {reward1:.1f}\\nStep: {step}')
        ax1.axis('off')

        # Trained agent display
        ax2.clear()
        ax2.imshow(env2.render())
        ax2.set_title(f'Trained Agent\\nReward: {reward2:.1f}\\nStep: {step}')
        ax2.axis('off')

        plt.tight_layout()
        # Update the displayed figure
        display(fig)


      obs1, obs2 = next_obs1, next_obs2

      if (term1 or trunc1) and (term2 or trunc2):
        break

    env1.close()
    env2.close()
    plt.close(fig) # Close the figure after the comparison


    print(f"\n📊 Comparison Results:")
    print(f"Random Agent - Steps: {step+1}, Reward: {reward1}")
    print(f"Trained Agent - Steps: {step+1}, Reward: {reward2}")
    print(f"Performance Improvement: {reward2 - reward1:.1f} points")
```
